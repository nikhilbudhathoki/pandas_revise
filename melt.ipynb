### Melt

```python
import pandas as pd
```

```python
var=pd.DataFrame({'days':[1,2,3,4,5,6,7],'maths':[33,22,45,66,44,67,77],'eng':[77,55,88,44,96,44,33]})
```

```python
var
```

```python
pd.melt(var)
```

```python
pd.melt(var,id_vars=['days'])
```

```python
pd.melt(var,id_vars=['days'],var_name='subject',value_name='marks')
```
