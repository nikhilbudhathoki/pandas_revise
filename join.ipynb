```python
import pandas as pd
```

```python
ph=pd.DataFrame({'A':[1,2,3,4],'B':[22,33,44,55]})
ph2=pd.DataFrame({'C':[1,2,3,46],'D':[2,33,44,5]})
```

```python
ph.join(ph2)
```

```python
ph2.join(ph)
```

```python
ph=pd.DataFrame({'A':[1,2,3,4],'B':[22,33,44,55]},index=['a','b','c','d'])
ph2=pd.DataFrame({'C':[1,2,3,46],'D':[2,33,44,5]},index=['a','b','c','d'])
```

```python
ph.join(ph2)
```

```python
ph=pd.DataFrame({'A':[1,2,3,4],'B':[22,33,44,55]},index=['a','b','c','d'])
ph2=pd.DataFrame({'C':[1,2],'D':[2,33]},index=['a','b'])
```

```python
ph.join(ph2)
```

```python
ph.join(ph2,how='right')
```

```python
ph.join(ph2,how='left')
```

```python
ph.join(ph2,how='outer')
```

```python
ph.join(ph2,how='inner')
```

```python
ph=pd.DataFrame({'A':[1,2,3,4],'B':[22,33,44,55]},index=['a','b','c','d'])
ph2=pd.DataFrame({'B':[1,2],'D':[2,33]},index=['a','b'])
```

```python
ph.join(ph2,how='outer',lsuffix='.one')
```

```python
ph.join(ph2,how='outer',lsuffix='.one',rsuffix='two')
```

```python
ph.join(ph2,how='outer',lsuffix='.one',rsuffix='.one')
```
